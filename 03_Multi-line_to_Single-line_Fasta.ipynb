```python
def fasta_to_single_line(fasta_path):
    with open(fasta_path, 'r') as f:
        header_found = False
        for record in f:
            record = record.strip()
            if record.startswith('>'):
                if header_found:
                    print()
                print(record)
                header_found = True
            else:
                print(record, end='', flush=True)
        if header_found and not record.startswith('>'):
            print()

file_path = 'multiline_input.fasta'
fasta_to_single_line(file_path)
```
